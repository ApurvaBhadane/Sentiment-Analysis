```python
import numpy as np
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
import plotly
import plotly.graph_objs as go
import plotly.express as px
import seaborn as sns
```

```python
data_duplicates = pd.read_csv("/content/apps.csv.zip")
data_duplicates .head()
```

```python
apps = data_duplicates .drop_duplicates()
print("Total number of Apps: ",apps['App'].count())
apps.head()
```

```python
ch_remove = ('+','$','M',',')
col_clean = ('Size','Installs','Price')
for col in col_clean:
    for char in ch_remove:
        apps[col] = apps[col].str.replace(char,'')
    apps[col] = pd.to_numeric(apps[col])
```

```python
n_categories = len(set(apps['Category']))
print(" Number of categories: ",n_categories)
n_apps_in_category = apps['Category'].value_counts().sort_values(ascending=False)
data = go.Bar(
x = n_apps_in_category.index,
y = n_apps_in_category.values
)
fig = go.Figure(data=[data])
fig.show()

```

```python
print("Average App rating: ",apps['Rating'].mean())
```

```python
from re import X
data = [go.Histogram(
    x = apps['Rating'],
    xbins ={'start' : 1,'size':0.1,'end':5})]
avg_rating = apps['Rating'].value_counts().mean() # Removed extra indent here
layout = {'shapes': [{
    'type':'line',
    'x0': avg_rating,
    'y0':0,
    'x1': avg_rating,
    'y1':1000,
    'line': {'dash': 'dashdot'}
}]}
plotly.offline.iplot({'data':data,'layout':layout})
```

```python
sns.set_style("darkgrid")
import warnings
warnings.filterwarnings("ignore")
large_categories = apps.groupby('Category').filter(lambda x: len(x) > 250).reset_index()
paid_app = apps[apps['Type'] == 'Paid']
plt1 = sns.jointplot(x = 'Size', y = 'Rating', data = large_categories)
plt1 = sns.jointplot(x = 'Price', y = 'Rating', data = paid_app)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(15, 8)
popular_app_cats = apps[apps.Category.isin(['GAME','FAMILY','MEDICAL','TOOLS','FINANCE','LIFESTYLE'])]
ax = sns.stripplot(x = 'Price', y = 'Category', data = popular_app_cats,jitter = True, linewidth = 1  ,hue = 'Category')
ax.set_title('App pricing trend across categories')
apps_above_200 = popular_app_cats[['Category','App','Price']][popular_app_cats['Price'] > 200]
apps_above_200
```

```python
tr0 = go.Box (y=apps['Installs'][apps['Type'] == 'Paid'], name = 'Paid')
tr1 = go.Box (y=apps['Installs'][apps['Type'] == 'Free'], name = 'Free')
layout = go.Layout(title = "Number of downloads of paid apps vs. free apps",yaxis = dict(type = 'log',autorange = True))
data = [tr0,tr1]
plotly.offline.iplot({'data':data,'layout':layout})
```

```python
reviews_df = pd.read_csv("/content/user_reviews.csv (2).zip")
merged_df = pd.merge(apps,reviews_df,on = 'App',how ='inner')
merged_df = merged_df.dropna(subset=["Sentiment","Translated_Review"])

sns.set_style("ticks")
fig, ax = plt.subplots()
fig.set_size_inches(11,8)

ax = sns.boxplot(x ="Type" , y = "Sentiment_Polarity", palette = "Blues", data = merged_df)
ax.set_title("Sentiment Polarity Distribution")
```
